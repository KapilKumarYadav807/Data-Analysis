President General Polls Dataset

## Get the data

```python
# imports panda library as pd
import pandas as pd
# Defining a variable poll_url having the URL of the CSV file
poll_url = 'http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'
# uses read_csv() function from pandas to read the data from the CSV file 
polls = pd.read_csv(poll_url)
```

```python
# Displaying the contents of the 'polls' DataFrame
polls
```

```python
# Uses sort_values() method of the 'polls' DataFrame to sort the rows based on the values in the 'state' column in-place
polls.sort_values('state', inplace=True)

# Display the first 5 rows of the sorted 'polls' DataFrame
polls.head()
```

Use Tab completion and tooltips

```python
# Uses sort_values() method of the 'polls' DataFrame to sort the rows based on the values in the 'state' column in descending order
polls.sort_values('state', ascending=False)

# Display the first 5 rows of the sorted 'polls' DataFrame
polls.head()
```

```python
## Clean the data
```

### Keep only the rows with "now-cast" in the type column

```python
# Using query method to filter rows in the 'polls' DataFrame where the type column is equal to now-cast
polls = polls.query('type=="now-cast"')
# Display the first 5 rows of the 'polls' DataFrame
polls.head()
```

### Rename voting columns

```python
# Use the rename method to rename columns in the 'polls' DataFrame
polls.rename(columns={'rawpoll_clinton':'Clinton_pct',
                      'rawpoll_trump':'Trump_pct'},inplace=True)
# Display the first 5 rows of the 'polls' DataFrame
polls.head()
```

### Convert date columns to datetime objects

```python
# Create a list date_cols containing the names of columns
date_cols = ["startdate","enddate"]
# Convert specified date columns to datetime format using apply() and pd.to_datetime()
polls[date_cols] = polls[date_cols].apply(pd.to_datetime)
# Display the first 5 rows of the 'polls' DataFrame
polls.head()
```

### Try two plots

```python
%%time
# %%time magic command to measure the execution time

# Plot a line chart, 'enddate' column is specified as the x-axis, and the y-axis includes the columns 'Clinton_pct' and 'Trump_pct'
polls.plot.line(x='enddate',y=['Clinton_pct','Trump_pct'])
```

```python
%%time
# %%time magic command to measure the execution time

# Query the 'polls' DataFrame to filter rows where the state column is U.S
# Plot a line chart with 'enddate' on the x-axis and 'Clinton_pct' and 'Trump_pct' on the y-axis
polls.query('state == "U.S."') \
    .plot(x='enddate',y=['Clinton_pct','Trump_pct'])
```
